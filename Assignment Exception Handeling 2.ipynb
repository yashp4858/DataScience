```python
#Q3
#Arithmetic Error is thrown when an error occur while performing mathematical operatios.
#ZeroDivisionError example 

try:
    a=5/0
except ZeroDivisionError as e:
    print(e)
```

```python
#OverFlowError example
try:
    a = 5.0
    for i in range(1,1000):
        a = a**i
except OverflowError as e:
    print(e)
```
